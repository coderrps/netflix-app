```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
netflix = pd.read_csv('netflix.csv')
```

```python
# netflix['index'] = netflix['index'].astype(int)
```

```python
netflix.head()
```

```python
for columns in netflix:
    netflix[columns] = netflix[columns].fillna('')
```

```python
netflix.head()
```

```python
combine_columns = netflix['director'] + ' ' + netflix['cast'] + ' ' + netflix['listed_in'] + ' ' +netflix['description']
```

```python
print(combine_columns)
```

```python
vectorizer = TfidfVectorizer()
feature_vector = vectorizer.fit_transform(combine_columns)
print(feature_vector)
```

```python
similarity = cosine_similarity(feature_vector)
print(similarity)
```

```python
movie_name = input("Enter the movie name \n")
```

```python
title_list = netflix['title'].tolist()
print(title_list)
```

```python
find_match = difflib.get_close_matches(movie_name, title_list)
print(find_match)
```

```python
close_match = find_match[0]
```

```python
# movie_index = netflix[netflix.title == close_match]['index'].values[0]
# print(movie_index)
movie_index = netflix[netflix['title'] == close_match]['index'].values
if len(movie_index) > 0:
    movie_index = movie_index[0]
else:
    print("Movie not found,")
    exit()
```

```python
similarity_score = list(enumerate(similarity[movie_index]))
```

```python
sort_movie = sorted(similarity_score, key=lambda x:x[1], reverse=True)
print(sort_movie)
```

```python
i=1
for movies in sort_movie:
    index = movies[0]
    title_from_index = netflix[netflix.index==index]['title'].values[0]
    
    if(i<=20):
        print(i, title_from_index)
    i+=1
```
